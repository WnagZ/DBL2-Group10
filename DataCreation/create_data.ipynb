```python
import csv
import pandas as pd
import os
import shutil
```

```python
dir_path = './sorted_data'
new_root_dir = './sorted_data_2'

for dir_path, dir_names, file_names in os.walk(dir_path):
    for file_name in file_names:
        if file_name.split('-')[2] == 'metropolitan':
            year = dir_path.split("\\")[1].split("-")[0]
            new_dir_path = os.path.join(new_root_dir, year)

            # Create new directory if it doesn't exist
            if not os.path.exists(new_dir_path):
                os.makedirs(new_dir_path)

            old_file_path = os.path.normpath(os.path.join(dir_path, file_name))
            new_file_path = os.path.normpath(os.path.join(new_dir_path, file_name))

            # Move the file to the new directory
            shutil.move(old_file_path, new_file_path)
```

```python
dir_path = './sorted_data_2'
files = []

subfolders = [ f.path.replace("\\", "/") for f in os.scandir(dir_path) if f.is_dir() ]
for file in subfolders:
    year =  file.split('/')[2]
    print(file)
    dfs = []
    for (dirpath, dirnames, filenames) in os.walk(file):
        for x in filenames:
            df = pd.read_csv(f'{file}//{x}')
            dfs.append(df)
    df = pd.concat(dfs, ignore_index=True)
    df.to_csv(f'{year}.csv')
    print("done")
```

```python
dfs = []
for (dirpath, dirnames, filenames) in os.walk('./csv_files'):
    for x in filenames:
        df = pd.read_csv(f'./csv_files/{x}')
        dfs.append(df)
```
